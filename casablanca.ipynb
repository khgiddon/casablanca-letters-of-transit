```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.ticker as mtick
import matplotlib as mpl
import numpy as np

mpl.rc('font', family='Arial')

# P(D) should be a list of probabilities from 0 to 1
P_D = np.arange(0, 1.01, 0.01)

# P(S|D) = 1
P_S_given_D = 1

# P(S|~D) values
P_S_given_not_D_values = [0.1, 0.25, 0.5, 1]

# Colors for the lines
colors = ['blue', 'green', 'red', 'purple']

# Create graph
fig, ax = plt.subplots(figsize=[3.5, 3], dpi=300)

# Set backgrounds to light grey
fig.patch.set_facecolor('#f0f0f0')
ax.set_facecolor('#f0f0f0')

# Remove box (spines)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Set major gridlines
ax.grid(True, color='#cfcfcf',linewidth=0.75)

for i, P_S_given_not_D in enumerate(P_S_given_not_D_values):
    # Calculate P(D|S) using Bayes's formula
    P_D_given_S = [(P_D[j] * P_S_given_D) / ((P_D[j] * P_S_given_D) + ((1 - P_D[j]) * P_S_given_not_D)) for j in range(len(P_D))]

    # Plotting the graph
    ax.plot(P_D, P_D_given_S, label=f'P(S|~D)={int(P_S_given_not_D*100)}%', linewidth=1.5, alpha=.4, color=colors[i])
    
ax.tick_params(axis = 'both', which = 'major', labelsize = 5)
ax.tick_params(axis='x', rotation=0)
ax.axhline(y=0, color='#414141', linewidth=0.75, alpha=.35)
ax.axvline(x=0, color='#414141', linewidth=0.75, alpha=.35)

# Axis labels
ax.set_ylabel('Probability\nthat Ugarte\nsaid\n"de Gaulle,"\ngiven that\nthe subtitle\n says de Gaulle', rotation=0, size=5.5)
ax.yaxis.set_label_coords(-0.15,0.43)
ax.set_xlabel('Probability that Ugarte said "de Gaulle"',rotation=0,size=5.5)
# Format
ax.tick_params(axis='both', length=0, pad=2)
ax.xaxis.set_major_formatter(mtick.PercentFormatter(1.0))
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

# Set y-axis max to 100%
ax.set_ylim(0, 1)

# Adding legend, title, and labels
# Adding legend
legend = ax.legend(loc='lower right', prop={'size': 5}, title='Probability the\nsubtitle says\n"de Gaulle" if Ugarte\n did not say de Gaulle', title_fontsize=5.5)
legend.get_frame().set_alpha(0.5)
ax.text(x = -.3, y = 1.11, s = "Who signed the letters of transit in Casablanca?",
               fontsize = 10.5, weight = 'bold', alpha = .75)
ax.text(x = -0.3, y = 1.05, s = "A not-so-necessary Bayesian approach to a not-so-important mystery",
               fontsize = 7, alpha = .75)

# Footer
ax.text(x = -0.23, y = -0.14, 
        s = """© Kyle Giddon, 2024, www.kylegiddon.com""" + ''.join(' ' for _ in range(92)) + "Images are sourced from Wikimedia Commons and are in the public domain.",
        fontsize = 3, 
        color = '#f0f0f0', 
        bbox=dict(facecolor='grey', edgecolor='none', pad=2),
        transform=ax.transAxes, ha='left', va='top')

plt.show()
```
